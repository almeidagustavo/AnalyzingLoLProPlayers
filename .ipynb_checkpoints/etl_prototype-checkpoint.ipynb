```python
baseUrl = 'https://kr.api.riotgames.com/lol'
apiKey = 'RGAPI-7c4a3d4e-23d3-4868-b2b0-81c1e94541a1'
headers = {"Accept-Language": "pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7", "Accept-Charset": "application/x-www-form-urlencoded; charset=UTF-8"}
type(headers)
```

```python
def getChallengers(baseUrl = baseUrl,
                   apiKey = apiKey,
                   headers = headers,
                   endPoint = "/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5"):
    headers["X-Riot-Token"] = apiKey
    res = requests.get(baseUrl + endPoint, headers=headers)
    return json.loads(res.text)
```

```python
def getAccountId(summonerId,
                     baseUrl = baseUrl,
                     apiKey = apiKey,
                     headers = headers,
                     endPoint = "/summoner/v4/summoners/{encryptedSummonerId}"):
    headers["X-Riot-Token"] = apiKey
    res = requests.get(baseUrl + endPoint, headers=headers)
    return json.loads(res.text)
```

```python
def getRankedMatches(summonerId,
                     baseUrl = baseUrl,
                     apiKey = apiKey,
                     headers = headers,
                     endPoint = "/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5"):
    headers["X-Riot-Token"] = apiKey
    getAccountId(summonerId)
    res = requests.get(baseUrl + endPoint, headers=headers)
    return json.loads(res.text)
```

```python
def getDB():
    challengers = getChallengers()
    data = {
        "summonerId": [], 
        "summonerName": [],  
        "win": [], 
        "losses": [],
        "rank": [],
        "leaguePoints": [] # Our target
    }
    for challenger in challengers['entries']:
        print(challenger['summonerId'])
```
